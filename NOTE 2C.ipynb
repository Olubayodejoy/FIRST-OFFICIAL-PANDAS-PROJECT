•  US Census Data

Task: Load the dataset, filter rows based on demographic information, and create new columns for income brackets.
    
 

```python
import pandas as pd
import seaborn as sns
US_Census_df = pd.read_csv('C:\\Users\\HP\\Documents\\adult.csv')
US_Census_df
```

```python
US_Census_df.info()
```

```python

filtered11_df = US_Census_df[(US_Census_df['Age'] >= 30) & (US_Census_df['Race'] == 'White')]
filtered11_df
```

```python
US_Census_df['Income'] = US_Census_df['Income'].str.strip()
def  income_bracketss(salary):
    if salary == '>50K':
        return 'High'
    
    else:
        return 'Low'
    
US_Census_df['Real Income Brackets'] = US_Census_df['Income'].apply(income_bracketss)
US_Census_df
```

```python
US_Census_df['Workclass'] = US_Census_df['Workclass'].str.strip()
def  income_brackets(wclass):
    if wclass == 'Private':
        return 'Really High'
    elif wclass == 'State-gov':
        return  'High'
    elif wclass == 'Self-emp-inc':
        return  'Average'
    elif wclass == 'Self-emp-not-inc':
        return 'Okay'
    else:
        return 'Low'
    
US_Census_df['Income Brackets'] = US_Census_df['Workclass'].apply(income_brackets)
US_Census_df
```

•  Student Performance Data

      Task: Load the dataset, filter students by gender and performance level, and create new columns for grade categories.


```python
import pandas as pd
import seaborn as sns
student_Performance_df = pd.read_csv('C:\\Users\\HP\\Documents\\student-por.csv')
student_Performance_df
```

```python
filtered_grade7 = student_Performance_df[student_Performance_df['sex'] == 'F']
filtered_grade7
```

```python
filtered_grade8 = student_Performance_df[student_Performance_df['sex'] == 'M']
filtered_grade8
```

```python

def grade_category(score):
    if score >= 20:
        return 'A'
    elif score >= 15:
        return 'B'
    elif score >= 10:
        return 'C'
    elif score >= 5:
        return 'D'
    else:
        return 'F'

student_Performance_df['G1_grade'] = student_Performance_df['G1'].apply(grade_category)
student_Performance_df['G2_grade'] = student_Performance_df['G2'].apply(grade_category)
student_Performance_df['G3_grade'] = student_Performance_df['G3'].apply(grade_category)

student_Performance_df
```
